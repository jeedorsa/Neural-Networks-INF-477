## 2. *Transfer Learning*

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2017/05/31130754/transfer-learning.jpeg" , style="float: right;" />

En esta sección se trabajará con el dataset trabajado anteriormente, CIFAR [3], pero en su versión más fina, en el cual se presentan 100 tipos distintos de categorías a clasificar la imagen (mutuamente excluyente), no 10 como se usó en las actividades anteriores. La estructura es la misma, son 60000 imágenes RGB de 32 $\times$ 32 píxeles separados en 50 mil de entrenamiento y 10 mil de pruebas.  
Aquí se experimentará con el concepto de *transfer learning* el cual consta en transferir conocimiento de un dominio fuente (*source domain*) a un dominio objetivo (*target domain*). En redes neuronales existen muchas representaciones de esto, en común consta en pre inicializar los pesos de la red de alguna manera que no sea con distribuciones de manera aleatoria (*fine tunning*). También está lo que es utilizar una representación generada a través de otra red entrenada con muchos datos, esto es tomar la red y "*congelar*" sus primeras capas para tomar esta representación y no entrenar esos pesos, lo que realizaremos en esta sección. 

```python
import keras
from keras.datasets import cifar100
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten,BatchNormalization
from keras.layers import Conv2D, MaxPooling2D
import numpy as np  
import os
import matplotlib.pyplot as plt
from keras.utils.np_utils import to_categorical   
from keras.layers import Dense, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Input
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
from keras.callbacks import LearningRateScheduler
import math
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta
from keras.applications import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.models import Model
```

```python

(x_train, y_train), (x_test, y_test) = cifar100.load_data()
num_classes = 100
total=len(x_train)
Xtr=x_train[:int(0.9*total)]  
Xv=x_train[int(0.9*total)::]
Xt=x_test
total1=len(y_train)
Ytr=y_train[:int(0.9*total1)]  
Yv=y_train[int(0.9*total1)::]
Yt=y_test
```

```python
print("Shape of training data:")
print(Xtr.shape)
print(Ytr.shape)
print("Shape of test data:")
print(Xt.shape)
print(Yt.shape)
print("Shape of validation data:")
print(Xv.shape)
print(Yv.shape)
```

```python
Ytr = to_categorical(Ytr, num_classes=100)
Yt = to_categorical(Yt, num_classes=100)
Yv= to_categorical(Yv, num_classes=100)
Xtr = Xtr.astype('float32')
Xt = Xt.astype('float32')
Xv= Xv.astype('float32')
Xtr /= 255
Xt /= 255
Xv /= 255
print("Shape of training data:")
print(Xtr.shape)
print(Ytr.shape)
print("Shape of test data:")
print(Xt.shape)
print(Yt.shape)
print("Shape of validation data:")
print(Xv.shape)
print(Yv.shape)
```

**a**) Entrene una red neuronal convolucional como se presenta en el código a continuación durante 15 *epochs*, realizando un gráfico de evolución de la función de pérdida y de la exactitud del algoritmo (*accuracy*) sobre ambos conjuntos, entrenamiento y pruebas. Comente sobre el tiempo de ejecución de este entrenamiento. Reporte el *accuracy* del modelo final sobre el conjunto de pruebas. 

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), padding='same',input_shape=Xtr.shape[1:],activation='relu'))
model.add(Conv2D(32, (3, 3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(1024,activation='relu'))
model.add(Dense(100, activation='softmax'))
sgd =  SGD(lr=0.001)   
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=sgd)
history = model.fit(Xtr, Ytr, batch_size=128, epochs=15, verbose=1, validation_data=(Xv,Yv))
model.summary()
```

```python
plt.figure(1,figsize=(12,8))
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])
plt.legend(('Val Loss ', 'Loss Train' ))
plt.title('GRAFICA DE LOSS')
plt.ylabel('MSE')
plt.xlabel('Epochs')
```

```python
validate = model.evaluate(Xv, Yv)
print("Loss validacion:")
print(validate[0])
print("Accuracy :")
print(validate[1]*100)
```

En este caso aunque mantenemos el mismo numero de imagenes tenemos un numero superior de clases, esto conlleva tener un numero mayor de parametros libres para el entrenamiento, y al parecer nos esta incidiendo en tener un accuracy bastante inferior, pues la funcion de costo nunca logra tener un minimo deseado y logramos tan solo una efectividad de clasificacion con el set de datos de testeo de 1%.

**b**) Debido al comportamiento de las curvas de entrenamiento, claramente se ve que se necesita un regularizador. Experimente utilizando Dropout con una tasa de 0.25 en las tandas convolucionales, elija donde situarlo, luego de la primera convolución, después de la segunda, solamente después del *pooling*, en todas o alguna forma que le parezca conveniente, de argumentos de ello. La idea es que se forme una idea de dónde conviene colocar el regularizador y porqué.

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), padding='same',input_shape=Xtr.shape[1:],activation='relu'))
model.add(Dropout(0.25))
model.add(Conv2D(32, (3, 3),padding='same',activation='relu'))
model.add(Dropout(0.25))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(100, activation='softmax'))
sgd =  RMSprop(lr=0.00001)   
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=sgd)
history = model.fit(Xtr, Ytr, batch_size=32, epochs=15, verbose=2, validation_data=(Xv,Yv))
model.summary()
```

```python
plt.figure(1,figsize=(12,8))
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])
plt.legend(('Val Loss ', 'Loss Train' ))
plt.title('GRAFICA DE LOSS DROPOUT EN DOS CAPAS ')
plt.ylabel('MSE')
plt.xlabel('Epochs')
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), padding='same',input_shape=Xtr.shape[1:],activation='relu'))
model.add(Dropout(0.25))
model.add(Conv2D(32, (3, 3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(1024,activation='relu'))
model.add(Dense(100, activation='softmax'))
sgd =  RMSprop(lr=0.00001)   
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=sgd)
history = model.fit(Xtr, Ytr, batch_size=32, epochs=15, verbose=2, validation_data=(Xv,Yv))
model.summary()
```

```python
plt.figure(1,figsize=(12,8))
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])
plt.legend(('Val Loss ', 'Loss Train' ))
plt.title('GRAFICA DE LOSS DROPOUT EN PRIMERA CAPA ')
plt.ylabel('MSE')
plt.xlabel('Epochs')
```

En el primer modelo que entrenamos hemos introducido dropout en ambas capas, esta herramienta aunque nos permite evitar el overfitting ese es un problema que en estos momentos no estamos afrontando y aunque nos da en el primer ejemplo un aumento de 9%, no creemos que es suficiente, uno de los problemas de tener dropout es que en algunos momentos da un problema de backpropagation, cuando solo aplicamos dropout sobre la primera capa, mejoramos casi el doble del accuracy con respecto al primer modelo, pero sigue siendo un nivel muy bajo comparado con las aplicaciónes que hemos realizado en el primer ejercicio donde alcanzabamos niveles de accuracy de 70-90%.

c) Una forma de hacer lo que se conoce como *transfer learning* es utilizar el conocimiento (los parámetros) aprendido por una red entrenada con millones de imágenes, y tomar estos parámetros como los pre entrenados. Para esto se utilizará el modelo VGG16 [7] proporcionado a través de la interfaz de *keras*. Visualice el modelo y sus 23 capas. Para esta instancia se utilizará todo lo aprendido por las capas convolucionales, es decir, se eliminan las capas densas del modelo y se agregan unas nuevas a ser entrenadas desde cero.  
*Recuerde normalizar los datos de la manera en que fue entrenado VGG* ¿Cuál es éste proceso?

```python
(x_train, y_train), (x_test, y_test) = cifar100.load_data()
total=len(x_train)
Xtr=x_train[:int(0.9*total)]  
Xv=x_train[int(0.9*total)::]
Xt=x_test
```

```python
import cv2
x=Xtr[3]
plt.imshow(x)
```

```python
for i in range(0,32):
  for j in range(0,32):
    x[i][j][0]=x[i][j][0]-103.939
    x[i][j][1]=x[i][j][1]-116.779
    x[i][j][2]=x[i][j][2]-123.68
plt.imshow(x)
```

```python
def normalizarvgg(x):
  l=len(x)
  for m in range (0,l):
    img=x[m]
    for i in range(0,32):
      for j in range(0,32):
        img[i][j][0]=img[i][j][0]-103.939
        img[i][j][1]=img[i][j][1]-116.779
        img[i][j][2]=img[i][j][2]-123.68
    x[m]=img
  return(x)
Xtr=normalizarvgg(Xtr)
Xt=normalizarvgg(Xt)
Xv=normalizarvgg(Xv)
```

```python
plt.imshow(Xtr[5])
```

```python
x_train_vgg = preprocess_input(Xtr)
x_test_vgg = preprocess_input(Xv)
x_val_vgg= preprocess_input(Xt)
input_tensor=Input(shape=x_train_vgg.shape[1:])
modelVGG = VGG16(weights='imagenet', include_top=False,input_tensor=input_tensor ) # LOAD PRETRAINED MODEL 
features_train = modelVGG.predict(x_train_vgg)
features_test = modelVGG.predict(x_test_vgg)
modelVGG.summary()
```

Uno de los requerimientos basicos cuando vamos a utilizar un modelo predefinido es que tenemos que ajustar nuestros datos al modelo que se va a evaluar, en este caso para la normalización no podemos utilizar el mismo metodo que veniamos aplicando  de llevar los valores de cada capa entre 0 y 1, en este caso debemos aplicar una resta de valores de pixel especificos a cada una de las capas individuales. Un numero que no deja de llamarnos la atención es el numero de parametros libre de entrenamiento es casi 28 veces superior al mejor modelo del punto 1, y 10 veces superior al modelo implementado en la sección anterior.
En el codigo ademas mostramos la imagen real, y como queda despues de la normalzación, para poder ver que o como deben ser las caracteristica s que se desean aprender.

En la nomalización se efectuan las siguientes operaciones en cada pixel:

**CAPA ROJA**=CAPA ROJA-103.939
**CAPA VERDE**= CAPA VERDE -116.779
**CAPA AZUL**= CAPA AZUL- 123.68

**d)** esta red agregando una capa densa de 1024 neuronas seguido de un dropout de 0.5, finalmente es necesario agregar la capa de clasificación para las 100 clases. Utilice la misma configuración del optimizador para que las comparaciones sean válidas. Entrene unicamente por 10 epochs y grafique las curvas de entrenamiento con respecto al modelo definido en a) o con regularización definido en b) ¿Qué sucede? Comente

```python
             
model1 = Sequential()
model1.add(Flatten(input_shape=features_train.shape[1:]))
model1.add(Dense(1024, activation='relu'))
model1.add(Dropout(0.5))
model1.add(Dense(100, activation='softmax'))
opti=SGD(lr=0.001)
model1.compile(optimizer=opti,loss='categorical_crossentropy', metrics=['accuracy'])
history1=model1.fit(features_train, Ytr,epochs=10, batch_size=128,verbose=1,validation_data=(features_test,Yv))
model1.summary()
                         
```

```python
plt.figure(1,figsize=(12,8))
plt.plot(history1.history['val_loss'])
plt.plot(history1.history['loss'])
plt.legend(('Val Loss ', 'Loss Train' ))
plt.title('GRAFICA DE LOSS DROPOUT EN DOS CAPAS ')
plt.ylabel('MSE')
plt.xlabel('Epochs')
```

En este caso partimos el entrenamiento de la red densa con la salida de las caracteriticas calculadas por el modelo VGG16, aunque si en los primeros puntos observamos un comportamiento que no cumplia con lo que deseabamos en este caso al parecer se comete alguna equivocacion por que la funcion de costo converge sobre un valor de 6, con un accuracy maximo de 8%. 

**e)** Agregue una capa de normalización (Batch Normalization [8]) de las activaciones en las capas densas, esto es, restar por la media del batch y dividir por la desviación estándar. Vuelva a entrenar el modelo con la misma configuración pero ahora por 15 epochs. Comente lo observado y compare las curvas de convergencia con los modelos anteriores ¿Por qué esto mejora a lo presentado en e)? Realice los mismos gráficos que en a) a través del número de epochs y comente sobre el tiempo de ejecución de este entrenamiento.

```python
model2 = Sequential()
model2.add(Flatten(input_shape=features_train.shape[1:]))
model2.add(BatchNormalization())
model2.add(Dense(1024, activation='relu'))
model2.add(BatchNormalization())
model2.add(Dropout(0.5))
model2.add(Dense(100, activation='softmax'))
opti=SGD(lr=0.001)
model2.compile(optimizer=opti,loss='categorical_crossentropy', metrics=['accuracy'])
history2=model2.fit(features_train, Ytr,epochs=15, batch_size=128,verbose=1,validation_data=(features_test,Yv))
model2.summary()
                        
```

```python
plt.figure(1,figsize=(12,8))
plt.plot(history2.history['val_loss'])
plt.plot(history2.history['loss'])
plt.legend(('Val Loss ', 'Loss Train' ))
plt.title('GRAFICA DE LOSS DROPOUT EN DOS CAPAS ')
plt.ylabel('MSE')
plt.xlabel('Epochs')
```

Cuando se aplica batchnormalization tenemos una mejora en el accurac logrando el maximo de todo este ejercicio llegando a un 15%, puede que en algun momento estemos fallando en algun aspecto respecto al entrenamiento pero la forma en que se comportan los algoritmos nos permite tener una compresion muy grande del uso de todos los hiperparametros.

**g)** ¿Cuándo podría ser útil y cuando no utilizar *transfer learning* o una red pre-entrenada?

Es util cuando tenemos un set de datos parecidos  a las condiciones con las que fue entrenada, o si es posible hacer una normalización de los datos para que pueda funcionar correctamente, no es útil cuando debemos partir desde cero, cuando nuestros datos no pueden ser pre-procesados  o queremos hacer una clasificación específica.
